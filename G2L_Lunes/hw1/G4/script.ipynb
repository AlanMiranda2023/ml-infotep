# **Punto 1: Análisis Univariado**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


%matplotlib inline

fileCsv = pd.read_csv('sample_data/california_housing_test.csv')

vars = ['housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households', 'median_income', 'median_house_value']
```

```python
for var in vars:
    plt.figure(figsize=(10,4))
    sns.histplot(fileCsv[var], kde=True, color='blue', bins=30)

    # Cálculo de medidas
    media = fileCsv[var].mean()
    mediana = fileCsv[var].median()
    moda = fileCsv[var].mode()[0]

    # Líneas para las medidas de centralidad
    plt.axvline(media, color='red', linestyle='dashed', linewidth=2, label=f'Media ({media:.2f})')
    plt.axvline(mediana, color='green', linestyle='dashed', linewidth=2, label=f'Mediana ({mediana:.2f})')
    plt.axvline(moda, color='pink', linestyle='dashed', linewidth=2, label=f'Moda ({moda:.2f})')

    plt.title(f'Histograma y Medidas de Centralidad para {var}')
    plt.xlabel(var)
    plt.ylabel('Frecuencia')
    plt.legend()
    plt.show()
```

```python
for var in vars:
    plt.figure(figsize=(7,2))
    sns.boxplot(x=fileCsv[var], color='yellow')
    plt.title(f'Boxplot para {var}')
    plt.xlabel(var)
    plt.show()
```

```python
for var in vars:
    plt.figure(figsize=(8,6))
    sns.kdeplot(fileCsv[var], shade=True, color='blue')
    plt.title(f'Gráfico de Densidad para {var}')
    plt.xlabel(var)
    plt.ylabel('Densidad')
    plt.show()
```

## **Punto 2: Análisis Bivariado**


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

fileCsv = pd.read_csv('sample_data/california_housing_test.csv')

varis = ['median_house_value', 'total_rooms', 'total_bedrooms']
```

```python
corr_matrix = fileCsv[varis].corr()

plt.figure(figsize=(6,4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Matriz de Correlación")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='total_rooms', y='median_house_value', data=fileCsv, color='teal')
plt.title("Relación: Total Rooms vs Median House Value")
plt.xlabel("Total Rooms")
plt.ylabel("Median House Value")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='total_rooms', y='total_bedrooms', data=fileCsv, color='purple')
plt.title("Relación: Total Rooms vs Total Bedrooms")
plt.xlabel("Total Rooms")
plt.ylabel("Total Bedrooms")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.regplot(x='total_rooms', y='median_house_value', data=fileCsv, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title("Regresión: Total Rooms vs Median House Value")
plt.xlabel("Total Rooms")
plt.ylabel("Median House Value")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.regplot(x='total_bedrooms', y='median_house_value', data=fileCsv, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title("Regresión: Total Bedrooms vs Median House Value")
plt.xlabel("Total Bedrooms")
plt.ylabel("Median House Value")
plt.show()
```
