```python
import pandas as pd
import matplotlib.pyplot as plt

Datos = pd.read_csv('/content/sample_data/california_housing_test.csv')

DatosUni = ['housing_median_age',
            'total_rooms',
            'total_bedrooms',
            'population',
            'households',
            'median_income',
            'median_house_value'
            ]
DatosBiv = ['housing_median_age',
            'total_rooms',
            'median_house_value'
            ]

# Generar gráficos univariados
for var in DatosUni:
    print(f"Media: {Datos[var].mean():}")
    print(f"Mediana: {Datos[var].median():}")
    print(f"Desviación estándar: {Datos[var].std():}")
    print(f"Varianza: {Datos[var].var():}")


    plt.hist(Datos[var], bins=50, edgecolor='black', alpha=0.7, color='#90EE90')
    plt.title(f'Histograma de {var}')
    plt.xlabel(var)
    plt.ylabel('Frecuencia')
    plt.show()



# Calcular y mostrar la matriz de correlación
AnalisisBiv = Datos[DatosBiv].corr()
print("Matriz de Correlación:")
print(AnalisisBiv)

# Generar gráficos bivariados (scatter plots)
for i in range(len(DatosBiv)):
    for j in range(len(DatosBiv)):
        plt.scatter(Datos[DatosBiv[i]], Datos[DatosBiv[j]], edgecolor='#afedea', alpha=0.5)
        plt.xlabel(DatosBiv[i])
        plt.ylabel(DatosBiv[j])
        plt.title(f'{DatosBiv[i]} vs {DatosBiv[j]}')
        print("")
        plt.show()
```
