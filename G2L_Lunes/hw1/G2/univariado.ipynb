```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('sample_data/california_housing_test.csv')
print(df.head())
```

```python
import numpy as np

def analisis_univariado(df, columna, n_bins=5):
    print(f"Análisis univariado de {columna}:")

    plt.figure(figsize=(8, 6))
    sns.histplot(df[columna], kde=True)
    plt.title(f"Histograma de {columna}")
    plt.show()

    bins = np.linspace(df[columna].min(), df[columna].max(), n_bins + 1)
    categorias = pd.cut(df[columna], bins, include_lowest=True)


    proporciones = categorias.value_counts(normalize=True)

    plt.figure(figsize=(8, 6))
    proporciones.sort_index().plot(kind='bar')
    plt.title(f"Diagrama de barras de {columna} (proporciones)")
    plt.xlabel("Categorías")
    plt.ylabel("Proporción")
    plt.show()

    media = df[columna].mean()
    mediana = df[columna].median()
    moda = df[columna].mode()[0]
    print(f"  Medidas de centralidad:")
    print(f"    Media: {media:.2f}")
    print(f"    Mediana: {mediana:.2f}")
    print(f"    Moda: {moda:.2f}")

    desviacion_estandar = df[columna].std()
    varianza = df[columna].var()
    rango = df[columna].max() - df[columna].min()
    print(f"  Medidas de dispersión:")
    print(f"    Desviación estándar: {desviacion_estandar:.2f}")
    print(f"    Varianza: {varianza:.2f}")
    print(f"    Rango: {rango:.2f}")

    print("-" * 40)
```

```python
variables_a_analizar = [
    'housing_median_age', 'total_rooms', 'total_bedrooms', 'population',
    'households', 'median_income', 'median_house_value'
]

for variable in variables_a_analizar:
    analisis_univariado(df, variable)
```
