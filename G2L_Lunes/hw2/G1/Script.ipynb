```python
import pandas as pd
import numpy as np

llamar_archivo = 'https://raw.githubusercontent.com/jrudascas/ml-infotep/main/Datasets/houses_medellin.csv'

"""
Carga el documento CSV desde una URL y muestra información sobre los datos.
"""
data = pd.read_csv(llamar_archivo)



"""
Muestra las 429 filas.
"""
data.head(429)
```

```python
"""
Muestra la forma del DataFrame.
"""
data.shape
```

```python
"""
Elimina filas con menos del 50% de datos no nulos.
"""
filtros_data = data.dropna(thresh=int(data.shape[1] * 0.5))
```

```python
"""
Muestra los datos filtrados.
"""
filtros_data
```

```python
"""
Muestra la nueva forma del DataFrame filtrado.
"""
filtros_data.shape
```

```python
"""
Calcula el número de filas antes y después del filtrado.
"""
filtros_antiguos = data.shape[0]
filtros_nuevos = filtros_data.shape[0]
```

```python
"""
Imprime la cantidad de filas eliminadas.
"""
print(f'Se eliminaron {filtros_antiguos - filtros_nuevos} filas')
```
