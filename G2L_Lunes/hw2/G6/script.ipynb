```python
import pandas as pd
import numpy as np

# Cargar documento desde la URL raw
Url = 'https://raw.githubusercontent.com/jrudascas/ml-infotep/main/Datasets/houses_medellin.csv'
datos = pd.read_csv(Url)

# Ver los primeros registros
datos.head(430)



```

```python
#Cálculo del umbral y lo multiplica por 0,2 (20%)
threshold = len(datos.columns) * 0.2 

#Eliminar las filas o columnas que contienen valores nulos
data_cleaned = datos.dropna(axis=0, thresh=threshold)

data_cleaned.head(430)
```
