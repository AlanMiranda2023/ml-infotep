**Definimos** función para verificar si está normalizado

```python
import pandas as pd
import numpy as np

def is_normalized(column):
  if column.isnull().all():
    return False
  return column.min() >= 0 and column.max() <= 1
```

Ejemplo de uso de la libreria Pandas

```python
import pandas as pd

serie_ejemplo = pd.Series([0.2,0.5,0.8,1.0,0.0])
print(is_normalized(serie_ejemplo))
```

Normalización de un DataFrame

```python
def mormalized_dataframe(df):

  df_normalized = df.copy()

  for col in df_normalized.columns:
    if not is_normalized(df_normalized[col]):
      col_min = df_normalized[col].min()
      col_max = df_normalized[col].max()

      if col_max != col_min:
        df_normalized[col] = (df_normalized[col] - col_min) / (col_max - col_min)
      else:
         df_normalized[col] = 0

  return df_normalized
```

Aqui creamos una copia del DataFrame original, recorremos las columnas y si alguna no estpa normalizada utiliza la fórmula: xnormalizado = x-min/max-min

Ejemplo de uso de normalización

```python
import pandas as pd

df_test = pd.DataFrame({
    'precio': [50, 100, 150, 200, 300],
    'metros_cuadrados': [30, 45, 60, 75, 90]
})

df_normalizado = mormalized_dataframe(df_test)
print(df_normalizado)
```
