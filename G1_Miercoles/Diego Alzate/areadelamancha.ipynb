**Calculando el area de la mancha**


```python
from PIL import Image
import matplotlib.pyplot as plt
import random

try:
    img = Image.open("mancha.png") #cargar la imagen
    ancho, alto = img.size #Obteniendo el ancho y alto de la imagen que se ingreso

    print(f"Ancho de la imagen: {ancho} píxeles") #Se imprime
    print(f"Alto de la imagen: {alto} píxeles")

    # Convertir la imagen (escala de grises) que el el modo L para asi facilitar el conteo de puntos dentro de la mancha
    img_gris = img.convert("L")

    # Umbral para definir qué se considera parte de la mancha (ajusta este valor según la imagen)
    threshold = 90

    # Inicializar el contador de puntos dentro de la mancha
    contador_de_punto = 0


    n = 100 #cantidad de puntos

    puntos = [] #variable para guardar la ubicación de los puntos aleatorios
    for _ in range(n):
        x = random.randint(0, ancho - 1)
        y = random.randint(0, alto - 1)
        puntos.append((x,y))

        # Obtener el valor del píxel en escala de grises
        pixel_valor = img_gris.getpixel((x, y))

        # Si el valor del píxel es menor que el umbral, el punto está dentro de la mancha
        if pixel_valor < threshold:
            contador_de_punto += 1

    print(f"Puntos dentro de la mancha (aproximado): {contador_de_punto}")

    #calculando el area
    a= (contador_de_punto/n)*(ancho*alto)
    print(f"Area de la mancha: {a}")

    # Mostrar la imagen con los puntos
    plt.imshow(img)
    plt.axis('off')

    #asignando o cargando los puntos en la ubicacion ramdo que asigno anteriormente
    for x,y in puntos:
      plt.plot(x,y, marker='o', markersize=2, color='red')

    plt.show()

except FileNotFoundError:
    print("Error: El archivo 'mancha.png' no se encuentra en el directorio actual.")
except Exception as e:
    print(f"Error al cargar o mostrar la imagen: {e}")
```
