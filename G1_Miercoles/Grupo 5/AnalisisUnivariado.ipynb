```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo de los datos
file_path = "sample_data/california_housing_test.csv"
df = pd.read_csv(file_path)

variables =  ["housing_median_age", "total_rooms", "total_bedrooms", "population", "households", "median_income", "median_house_value"]

# Definir el tamaño de las grficas
figsize = (10, 5)

# Crear gráficos organizados para cada variable
for column in variables:
    data_series = df[column]

    # Histograma
    plt.figure(figsize=figsize)
    plt.hist(data_series, bins=20, color='royalblue', edgecolor='black', alpha=0.7)
    plt.title(f"Distribución de {column}", fontsize=14)
    plt.xlabel(column, fontsize=12)
    plt.ylabel("Frecuencia", fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

    #Gráfico de Barras para Medidas de Centralidad
    central_values = {"Media": data_series.mean(), "Mediana": data_series.median(), "Moda": data_series.mode()[0]}
    plt.figure(figsize=figsize)
    plt.bar(central_values.keys(), central_values.values(), color=["red", "green", "blue"])
    plt.title(f"Medidas de Centralidad de {column}", fontsize=14)
    plt.ylabel("Valor", fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

    #Diagrama de Caja (Boxplot) - Medidas de Dispersión
    plt.figure(figsize=figsize)
    sns.boxplot(x=data_series, color="orange")
    plt.title(f"Dispersión de {column} (Boxplot)", fontsize=14)
    plt.xlabel(column, fontsize=12)
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.show()

```
