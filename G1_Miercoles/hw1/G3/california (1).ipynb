```python
import pandas as pd

# Cargar datos de california housing test
d_california = pd.read_csv('/content/sample_data/california_housing_test.csv')

# Mostrar las primeras filas
d_california.head()
```

```python
import matplotlib.pyplot as plt

# Lista de variables
variables = ["housing_median_age", "total_rooms", "total_bedrooms", "households", "median_income", "median_house_value"]

# Grafica de Histogramas
d_california[variables].hist(figsize=(12, 8), bins=30, edgecolor='black')
plt.suptitle("Histogramas de las variables seleccionadas")
plt.show()
```

```python
# Cálculo de medidas
estadisticas = d_california[variables].describe().T
estadisticas
```

```python
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.scatterplot(x=d_california["total_rooms"], y=d_california["median_house_value"], alpha=0.5)
plt.title("Relación entre total de habitaciones y valor de la casa")
plt.xlabel("Total de Habitaciones")
plt.ylabel("Valor de la Casa")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=d_california["total_bedrooms"], y=d_california["median_house_value"], alpha=0.5)
plt.title("Relación entre total de dormitorios y valor de la casa")
plt.xlabel("Total de Dormitorios")
plt.ylabel("Valor de la Casa")
plt.show()
```

```python
import numpy as np

plt.figure(figsize=(8, 6))
sns.heatmap(d_california[["median_house_value", "total_rooms", "total_bedrooms"]].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de Correlación")
plt.show()
```
