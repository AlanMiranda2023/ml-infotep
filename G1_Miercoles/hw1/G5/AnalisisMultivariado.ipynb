```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los datos
file_path = "/content/sample_data/california_housing_test.csv"
df = pd.read_csv(file_path)

variables = ["median_house_value", "total_rooms", "total_bedrooms"]

# Graficar la matriz de correlación
correlation_matrix = df[variables].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=1, square=True)
plt.title("Matriz de Correlación entre median_house_value, total_rooms y total_bedrooms")
plt.show()
#FIN  Graficar la matriz de correlación


# Diagramas de dispersión
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Scatter plot: median_house_value vs total_rooms
sns.scatterplot(data=df, x="total_rooms", y="median_house_value", alpha=0.5, ax=axes[0])
axes[0].set_title("Relación entre Total de Habitaciones y Valor de la Vivienda")
axes[0].set_xlabel("Total de Habitaciones")
axes[0].set_ylabel("Valor Mediano de la Vivienda")

# Scatter plot: total_rooms vs total_bedrooms
sns.scatterplot(data=df, x="total_rooms", y="total_bedrooms", alpha=0.5, ax=axes[1])
axes[1].set_title("Relación entre Total de Habitaciones y Total de Dormitorios")
axes[1].set_xlabel("Total de Habitaciones")
axes[1].set_ylabel("Total de Dormitorios")

plt.tight_layout()
plt.show()
```
